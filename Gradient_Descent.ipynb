### GRADIENT DESCENT

We will use Loss function to find the best fit. Loss is an error in predicted value of m and c. Here goal is to minize the error and we can use mean squared error function for calculating loss.

Suppose y is actual value and y(i) is predicted value, then the mean squared function is
    ((y-y(i))^2)/n, for all values of i= 0 to n
    
Gradient Descent is an iterative optimization algorithm to find the minimum of function(Loss).

L learning rate is multiplied in order to decide how much m and c sould be changed in every iteration.

The mean squared function is now partially differentiated in order to iterate the value of m and c unless the value becomes minimum

```python
# Importing Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv(r'C:\Users\AMKASERA2001\Desktop\Ind_House\train.csv')

X = data.iloc[:, 0]
Y = data.iloc[:, 1]
```

```python
# Initializing the data
m = 0
c = 0

L = 0.0001 # Learning Rate
itr = 1000 # Number of Iteration
n = float(len(X))
```

```python
# Performing Gradient Descent 
for i in range(itr): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)
```

```python
# Making predictions
Y_pred = m*X + c

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
